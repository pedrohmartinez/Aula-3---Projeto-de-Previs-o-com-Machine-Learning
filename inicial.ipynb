# Projeto Python IA: Inteligência Artificial e Previsões

### Case: Score de Crédito dos Clientes

Você foi contratado por um banco para conseguir definir o score de crédito dos clientes. Você precisa analisar todos os clientes do banco e, com base nessa análise, criar um modelo que consiga ler as informações do cliente e dizer automaticamente o score de crédito dele: Ruim, Ok, Bom

```python
# Analisar o 1 cliente
# Se o cliente é bom, ok ou ruim (nota de crédito dele = score de crédito)

# Passo a passo do projeto:
# Passo 0: Entender o desafio da empresa
# Passo 1: Importar a base de dados
# Passo 2: Tratar e preparar a base de dados para a inteligência artificial
# Passo 3: Criar o modelo de previsão ou modelo de IA -> poor/ruim, standart/ok ou good/bom
# Passo 4: Avaliar e escolher o melhor modelo
# Passo 5: Fazer novas previsões

# Anotações:
# Uma IA sempre é treinada pra fazer uma única coisa, por isso, deve-se separar a tabela em duas partes
# X = o que eu vou usar para fazer a previsão
# Y = o que eu vou prever (único resultado)
# Uma IA só entende números, por isso, deve-se tratar a base para codificar todos os valores string em number
# Entender o desafio da empresa é muito importante para saber quão refinada e precisa, as respostas/resultados da IA deverão ser
```

```python
# Passo 1: Importar a base de dados
import pandas as pd

tabela = pd.read_csv("clientes.csv")

# visualizar tabela
display(tabela)
# visualzar informações/dados da tabela
display(tabela.info())
```

```python
# Passo 2: Tratar e preparar a base de dados para a inteligência artificial
from sklearn.preprocessing import LabelEncoder

# criar codificadores para transformar strings em numbers
# profissao
codificador_profissao = LabelEncoder()
tabela["profissao"] = codificador_profissao.fit_transform(tabela["profissao"])
# mix_credito
codificador_mix_credito = LabelEncoder()
tabela["mix_credito"] = codificador_mix_credito.fit_transform(tabela["mix_credito"])
# comportamento_pagamento
codificador_comportamento_pagamento = LabelEncoder()
tabela["comportamento_pagamento"] = codificador_comportamento_pagamento.fit_transform(tabela["comportamento_pagamento"])

# score_credito
display(tabela)
```

```python
# Prever a coluna "score_credito"

# X é o que eu vou usar para fazer a previsão
x = tabela.drop(columns=["id_cliente", "score_credito"])

# Y é o que eu vou prever ()
y = tabela["score_credito"]

# Separar quais dados a IA vai usar para aprender e quais vai usar para testar se aprendeu
# importação do modelo train_test_split que faz essa separação automaticamente
from sklearn.model_selection import train_test_split

# o padrão de separação da base de dados train_test_split é 70/30, porém pode ser reajustado 
# train_test_split(x,y, test_size=0.3) reajustado para 30%
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, test_size=0.3)
```

```python
# Passo 3: Criar o modelo de previsão ou modelo de IA -> ruim, ok ou bom

# Modelos mais usados no mercado:
# Árvore de decisão -> RandomForest 
# KNN -> Nearest Neighbors -> Vizinhos Próximos

# Passo a passo da criação do modelo:
    # Passo 01: Importar a IA(modelo)
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

    # Passo 02: Criar a IA(modelo)
modelo_arvore_decisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

    # Passo 03: Treinar a IA(modelo)
modelo_arvore_decisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
```

```python
# Passo 4: Avaliar e escolher o melhor modelo

# previsão
previsao_arvore = modelo_arvore_decisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)

# comparar a previsão com o y_teste
from sklearn.metrics import accuracy_score # significa percentual de acerto

display(accuracy_score(y_teste, previsao_arvore))
display(accuracy_score(y_teste, previsao_knn))

# Melhor modelo = Árvore de decisão!

# Como melhorar o modelo:
# GridSearch -> realizar vários treinamentos/testes com partes e parâmetros diferentes do modelo na base de dados pra chegar na melhor combinação possível
# Ter mais dados na base de dados
# Alterar a porcentagem de dados de teste e treino
# Testar outros parâmetros
```

```python
# Passo 5: Fazer novas previsões

# importar nova base de dados
novos_clientes = pd.read_csv("novos_clientes.csv")

# codificadores
# profissao
novos_clientes["profissao"] = codificador_profissao.transform(novos_clientes["profissao"])
# mix_credito
novos_clientes["mix_credito"] = codificador_mix_credito.transform(novos_clientes["mix_credito"])
# comportamento_pagamento
novos_clientes["comportamento_pagamento"] = codificador_comportamento_pagamento.transform(novos_clientes["comportamento_pagamento"])

previsao = modelo_arvore_decisao.predict(novos_clientes)

# se quiser adicionar a previsão como uma coluna na tabela use isso:
novos_clientes["previsão"] = previsao

display(novos_clientes)

display(previsao)
```
